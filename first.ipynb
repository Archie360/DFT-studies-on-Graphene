```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
import phonopy
import yaml
import pickle
from phonopy.phonon.band_structure import get_band_qpoints_and_path_connections
plt.style.use(['science','notebook'])
plt.rcParams['text.usetex'] = False
csfont = {'fontname':'Times New Roman'}
```

```python
plt.figure(figsize=(10,6))

efermi = -2.0483

data = np.loadtxt('mono-graphene/graphene.bands.gnu')
k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))

gG1 = k[0]; K = k[30]; M = k[60]; gG2 = k[-1]



plt.subplot(1,2,1)
csfont = {'fontname':'Times New Roman'}
plt.axhline (0, c='gray', ls=':')
plt.axvline(K, c='gray')
plt.axvline(M, c='gray')
for band in range(len(bands)):
    plt.plot(k, bands[band , :]-efermi , c='b', lw=2)
    plt.ylabel('Energy (eV)')
    plt.xlim(gG1 , gG2)
    plt.xticks ([gG1 , K, M, gG2], ['$\Gamma$ ', 'K', 'M','$\Gamma$ '])
    plt.tick_params(axis='x', which='minor', bottom=False , top=False)
# plt.savefig('graphene-bands.png')

with open('graphene-0/first_order/band.pkl', 'rb') as f:
    d = pickle.load(f)

gG1 = np.array(d['distances'])[0,:][0]
M = np.array(d['distances'])[1,:][0]
K = np.array(d['distances'])[2,:][0]
gG2 = np.array(d['distances'])[2,:][-1]

plt.subplot(1,2,2)
csfont = {'fontname':'Times New Roman'}
for j in range(6):    
    for i in [0,1,2]:
        plt.plot(np.array(d['distances'])[i,:],np.array(d['frequencies'])[i,:,j], c='b',lw=2)

plt.xlim(0,gG2)
plt.ylabel('Frequencies (THz)')
# plt.grid(False)
plt.xticks([gG1,M,K,gG2],["$\Gamma$","M","K","$\Gamma$"])
# plt.yticks(fontsize=10)
plt.axvline(x=M)
plt.axvline(x=K)
plt.axhline(y=0, linestyle='--')

plt.tight_layout()
plt.savefig('dispersion')
```

```python
with open('graphene-0/first_order/gv.pkl', 'rb') as f:
    mesh_dict = pickle.load(f)
```

```python
qpoints = mesh_dict['qpoints']
weights = mesh_dict['weights']
frequencies = mesh_dict['frequencies']
eigenvectors = mesh_dict['eigenvectors']
group_velocities = mesh_dict['group_velocities']
```

```python
group_velocity_magnitude = np.linalg.norm(group_velocities, axis=2) 


plt.figure(figsize=(8, 6))
for i in range(group_velocity_magnitude.shape[1]):
    plt.scatter(frequencies[:, i], group_velocity_magnitude[:, i], label=f'Mode {i+1}', s=7)

plt.xlabel('Frequency (THz)')
plt.ylabel('$V_g$')
plt.title('Phonon Group Velocities')
plt.legend()
plt.show()
```
