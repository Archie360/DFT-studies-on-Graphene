```python
import matplotlib.pyplot as plt
import numpy as np
# plt.style.use('classic')
plt.style.use(['science','grid','notebook'])
plt.rcParams['text.usetex'] = False

```

```python
plt.grid(True)
plt.figure(figsize=(5,6))


efermi = -2.0483

data = np.loadtxt('graphene.bands.gnu')
k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))

gG1 = k[0]; K = k[30]; M = k[60]; gG2 = k[-1]

# plt.figure ()
plt.axhline (0, c='gray', ls=':')
plt.axvline(K, c='gray')
plt.axvline(M, c='gray')


for band in range(len(bands)):
    plt.plot(k, bands[band , :]-efermi , c='b', lw=2)
    plt.xlabel('k points')
    plt.ylabel('Energy (eV)')
    plt.xlim(gG1 , gG2)
    # plt.ylim (-3, 3)
    plt.xticks ([gG1 , K, M, gG2], ['$\Gamma$ ', 'K', 'M','$\Gamma$ '])
    plt.tick_params(axis='x', which='minor', bottom=False , top=False)
plt.savefig('graphene-bands.png')
```
