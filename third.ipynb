```python
p=1
plt.figure(figsize=(10,3))

for f in [f1,f2,f3,f4,f5]: 

    g = f['gamma'][30]
    g = np.where(g > 0, g, -1)
    lifetime = np.where(g > 0, 1.0 / (2 * 2 * np.pi * g), 0)
    len(lifetime)
    
    plt.subplot(1,5,p)

    for i in range(1,6): 
        plt.scatter(f['frequency'][:,i],lifetime[:,i], s=25, c='w', edgecolor='r', marker = 'o', lw = 1.2)
        plt.xlim(0,50)
        plt.ylim(1e-1, 1e3)
        plt.semilogy()
        plt.yticks(fontsize=10)
        plt.xticks(fontsize=10)
        # plt.minorticks_off()
        # plt.ylim(0,200)
    
    p+=1

plt.tight_layout()
plt.savefig("lifetimes")
```
